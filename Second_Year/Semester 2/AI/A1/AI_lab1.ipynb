***Tensor Shapes***

On performing operations on two or more tensors, they will need to be of the same shape - that is, having the same number of dimensions and the same number of cells in each dimension. For that, we have the `torch.*_like()` methods:


```python
x = torch.empty(2, 2, 3)
print(x.shape)
print(x)

empty_like_x = torch.empty_like(x)
print(empty_like_x.shape)
print(empty_like_x)

zeros_like_x = torch.zeros_like(x)
print(zeros_like_x.shape)
print(zeros_like_x)

ones_like_x = torch.ones_like(x)
print(ones_like_x.shape)
print(ones_like_x)

rand_like_x = torch.rand_like(x)
print(rand_like_x.shape)
print(rand_like_x)
```

## Task 6 -- An application: Bertrand's paradox (probability) 

   Consider the [Bertrand paradox](https://en.wikipedia.org/wiki/Bertrand_paradox_(probability)).

   For each of the three cases generate a sample of 100 elements (chords) and compute the ratio of the chords bigger than the side of the equilateral triangle inscribed in the circle. 
   
   The circle radius is equal with 1 and the set of chords will contain the cartesian coordinates of the edges of the chords. The origin of the coordinate system is in the origin of the circle.   

   Draw using matplotlib each set of chords for each case respectively.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
SIDE = np.sqrt(3) # the size of a triangle inscribed in a circle of radius 1
RADIUS = 1 # circle radius
```

```python
# A function that returns the euclidean distance between 2 points (chords is a tuple of points)
def get_chord_length(chord):
    return np.sqrt((chord[1][0] - chord[0][0])**2 + (chord[1][1] - chord[0][1])**2)
```

```python
# A function that returns an array that represents the coordinates of the point with given radius and angle
def get_point(radius,angle):
    return radius * np.array([np.cos(angle), np.sin(angle)])
```

```python
# A function that computes the wanted ratio
def ratio(sample):
    count = 0
    for chord in sample:
        if get_chord_length(chord) > SIDE:
            count += 1

    return count / len(sample)
```

```python
# Define functions to compute the chord length for each case
# in each chase we return a chords as a pair of tuples (first_point, last_point)
# Case 1: random chord
def first_case_chord():
    theta1, theta2 = np.random.uniform(0, 2 * np.pi, size=2) # we choose 2 random angles on the circle 
    point1 = get_point(1,theta1)
    point2 = get_point(1,theta2)
    return np.array([point1, point2])
```

```python
# Case 2: chord perpendicular on radius
def second_case_chord():
    theta = np.random.uniform(0, 2 * np.pi) # choose a random angle
    point = get_point(1,theta)#random radius
    random_point = point * np.random.rand() # we take a rondom point on the radius
    perpen_vector = np.array([point[1], -point[0]]) # the perpendicular vector for (a,b) is the vector (b,-a)
    perpen_vector /= np.linalg.norm(perpen_vector) # we normalize the vector it will have length 1
    d_random_point = np.linalg.norm(random_point) # the distance from the origin to the chosen random point
    d_perpendicular_vect = (RADIUS- d_random_point ** 2) ** 0.5
    # Luam vectorul care este perpendicular pe raza si il scalam in functie de distanta potrivita si adunam si scadem la punctul aleatoriu ales pe raza cercului
    # astfel obtinem cele 2 puncte ale corzii
    return np.array([random_point - perpen_vector * d_perpendicular_vect, random_point + perpen_vector * d_perpendicular_vect])
```

```python
# Case 3: chord that passes through the inscribed triangle
def third_case_chord():
    random_point_for_radius = np.random.rand() ** 0.5 # we choose a random point for the inscribed circle of the triangle 
    theta = np.random.uniform(0, 2 * np.pi) # choose a random angle
    # next, we get the point that is on the distance of the inscribed circle in the triangle and on the angle theta
    random_point = get_point(random_point_for_radius,theta)
    
    perpen_vector = np.array([random_point[1], -random_point[0]])
    perpen_vector /= np.linalg.norm(perpen_vector)
    d_perpendicular_vect = (RADIUS - random_point_for_radius ** 2) ** 0.5
    
    return np.array([random_point - perpen_vector * d_perpendicular_vect, random_point + perpen_vector * d_perpendicular_vect])
```

```python
# Generate 100 chords for each case and compute the ratio
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 4))
circle = plt.Circle((0, 0), 1, fill=False)
circle1 = plt.Circle((0, 0), 1, fill=False)
circle2 = plt.Circle((0, 0), 1, fill=False)
triangle1 = plt.Polygon([[0, 1], [3**0.5 * 0.5, -0.5], [-3**0.5 * 0.5, -0.5]], fill='r')
triangle2 = plt.Polygon([[0, 1], [3**0.5 * 0.5, -0.5], [-3**0.5 * 0.5, -0.5]], fill='r')
triangle3 = plt.Polygon([[0, 1], [3**0.5 * 0.5, -0.5], [-3**0.5 * 0.5, -0.5]], fill='r')

ax1.add_artist(circle)
ax1.add_artist(triangle1)
ax2.add_artist(circle1)
ax2.add_artist(triangle2)
ax3.add_artist(circle2)
ax3.add_artist(triangle3)

for ax in (ax1, ax2, ax3):
    ax.set_xlim([-1, 1])
    ax.set_ylim([-1, 1])
    ax.set_aspect('equal')
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_aspect(1)

ax1.set_title("Case 1")
ax2.set_title("Case 2")
ax3.set_title("Case 3")


random_chords_case1 = [first_case_chord() for _ in range(100)]
for chord in random_chords_case1:
    ax1.plot(chord[:, 0], chord[:, 1],'b-',lw=0.5)
random_chords_case2 = [second_case_chord() for _ in range(100)]
for chord in random_chords_case2:
    ax2.plot(chord[:, 0], chord[:, 1],'b-',lw=0.5)
random_chords_case3 = [third_case_chord() for _ in range(100)]
for chord in random_chords_case3:
    ax3.plot(chord[:, 0], chord[:, 1],'b-',lw=0.5)
    
ratio_chords_case1 = ratio(random_chords_case1)
ratio_chords_case2 = ratio(random_chords_case2)
ratio_chords_case3 = ratio(random_chords_case3)

print("Case 1 ratio:", ratio_chords_case1)
print("Case 2 ratio:", ratio_chords_case2)
print("Case 3 ratio:", ratio_chords_case3)

plt.show()
```
